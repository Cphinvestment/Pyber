```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = pd.read_csv("data/city_data.csv")
ride_data_to_load = pd.read_csv("data/ride_data.csv")

# Read the City and Ride Data
merged_df=city_data_to_load.merge(ride_data_to_load)
# Combine the data into a single dataset

# Display the data table for preview
merged_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
citygroup=merged_df.groupby(['city','type'])
rides=citygroup['ride_id'].count()
fare=citygroup['fare'].mean()
drivers=citygroup['driver_count'].mean()
drop_city=merged_df.drop_duplicates('city')
test=drop_city.groupby('city')


#separated dfs for city categories

rural = merged_df.loc[(merged_df['type'] == 'Rural')]
suburban = merged_df.loc[(merged_df['type'] == 'Suburban')]
urban = merged_df.loc[(merged_df['type'] == 'Urban')]

# #Build the scatter plots for each city types
rural2=rural.groupby(['city'])['fare'].mean()
urban2=urban.groupby(['city'])['fare'].mean()
suburban2=suburban.groupby(['city'])['fare'].mean()

plt.scatter(rural.groupby(['city']).count()['ride_id'],rural2,s=rural.groupby(['city'])['driver_count'].value_counts()*10,facecolors=("gold"),edgecolor="black")
plt.scatter(urban.groupby(['city']).count()['ride_id'],urban2,s=urban.groupby(['city'])['driver_count'].value_counts()*10,facecolors=("lightcoral"),edgecolor="black")
plt.scatter(suburban.groupby(['city']).count()['ride_id'],suburban2,s=suburban.groupby(['city'])['driver_count'].value_counts()*10,facecolors=("lightskyblue"),edgecolor="black")
# Incorporate the other graph properties
plt.title('Note : Size of Bubble Corresponds to Number of Drivers per City')
plt.xlabel('Number of Rides per City')
plt.ylabel('Average Fare per City')

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

total_type=merged_df[['type','fare']]
total_type=total_type.groupby('type').sum()
total_type['percentage']=(total_type['fare']/total_type['fare'].sum())*100
# Build Pie Chart
plt.pie(total_type['percentage'],explode = (0,0,0.3),labels=['Rural','Suburban','Urban'],autopct="%1.1f%%", shadow=True,colors=['gold','skyblue','salmon'],startangle=140)
plt.title('% of Total Fares by City Type')
plt.axis('equal')
plt.show()

# Save Figure


plt.savefig("../Images/totalfare.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Build Pie Chart

# Save Figure
ride_type=merged_df[['type','ride_id']]
ride_type=ride_type.groupby('type').count()
ride_type['percentage']=(ride_type['ride_id']/ride_type['ride_id'].sum())*100
# Build Pie Chart
plt.pie(ride_type['percentage'],explode = (0,0,0.3),labels=['Rural','Suburban','Urban'],autopct="%1.1f%%", shadow=True,colors=['gold','skyblue','salmon'],startangle=140)
plt.title('% of Total rides by City Type')
plt.axis('equal')
plt.show()

# Save Figure

plt.savefig("../Images/totalrides.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

# Build Pie Charts

# Save Figure
ride_type=merged_df[['type','driver_count']]
ride_type=ride_type.groupby('type').count()
ride_type['percentage']=(ride_type['driver_count']/ride_type['driver_count'].sum())*100
# Build Pie Chart
plt.pie(ride_type['percentage'],explode = (0,0,0.3),labels=['Rural','Suburban','Urban'],autopct="%1.1f%%", shadow=True,colors=['gold','skyblue','salmon'],startangle=140)
plt.title('% of Total Drivers by City Type')
plt.axis('equal')
plt.show()

# Save Figure

plt.savefig("../Images/totaldrvers.png")
plt.show()
```
